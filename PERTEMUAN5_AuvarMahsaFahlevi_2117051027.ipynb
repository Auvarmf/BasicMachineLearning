# Tugas 

Evaluasi model algortima agglomerative clustering dari data latihan 2 dengan mencari nilai akurasinya

Hint: splitting data, modeling, accuracy score

Nilai Akurasi: 0.5079365079365079

```python
# melakukan pengambilan data
df = pd.read_csv("https://raw.githubusercontent.com/rahmantaufik27/dataset/main/study/student_score_pass.csv")
df
```

```python
# memilihan feature
# pada kasus ini kita akan menghapus data npm
df_training = df.drop(["npm"], axis = 1)

# konversi nilai pada kolom "pass_test" menjadi integer
# karena kita akan menggunakan data pass_test, maka dilakukan konversi terlebih dahulu
# hal ini dilakukan karena data pass_test merupakan string sehingga akan terjai error saat di olah, sehingga harus dilakukan konversi terlebih dahulu
df_training["pass_test"] = df_training["pass_test"].apply(lambda x: 1 if x == "pass" else 0)

# menghapus NaN values
df_training = df_training.dropna()
df_training
```

```python
# memisahkan data menjadi data latih dan data uji
X_train, X_test, y_train, y_test = train_test_split(df_training.drop(["pass_test"], axis=1), df_training["pass_test"], test_size=0.2, random_state=42)

# random_state adalah argumen opsional yang digunakan untuk menentukan seed atau biji secara acak yang digunakan dalam pembagian data.
# test_size digunakan untuk menentukan ukuran data test sebagai proporsi dari keseluruhan dataset.
# test_size = 0.2, maka 20% dari data akan digunakan sebagai data test, sementara 80% sisanya akan digunakan sebagai data training.
# ukuran test_size bisa disesuaikan dengan data yang dimiliki.
```

```python
# melakukan modeling Agglomerative Clustering
model = AgglomerativeClustering().fit(X_train)
model
```

```python
# melakukan prediksi pada data latih
train_prediksi = model.fit_predict(X_train)
train_prediksi
```

```python
# melakukan prediksi pada data uji
test_prediksi = model.fit_predict(X_test)
test_prediksi
```

```python
# menghitung nilai akurasi data latih
train_akurasi = accuracy_score(y_train, train_prediksi)
train_akurasi
```

```python
# menghitung nilai akurasi data uji
test_akurasi = accuracy_score(y_test, test_prediksi)
test_akurasi
```

Akurasi data tersebut sesuai dengan yang diharapkan walaupun terdapat sedikit perbedaan. perhitungan akurasi ini kemungkinan masih bisa ditingkatkan lagi untuk mendapatkan akurasi yang lebih baik.

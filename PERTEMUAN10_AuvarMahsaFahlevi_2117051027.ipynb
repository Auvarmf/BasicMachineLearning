## Library

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
from  sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/rahmantaufik27/dataset/main/study/weather_nb_dataset.csv")
df
```

## Preprocessing

```python
# pisahkan data target
X_train = df.drop('Play', axis='columns')
y_train = df['Play']
```



*   Kita ubah nilai outlook, temp, humidity, windy ke integer
*   hal ini ditunjukan untuk menyesuaikan nilai inputan modeling naive bayes yang ada di library python



```python
# buat dataframe baru dengan encoding data
numerics = LabelEncoder()
X_train['outlook_n'] = numerics.fit_transform(X_train['Outlook'])
X_train['Temp_n'] = numerics.fit_transform(X_train['Temp'])
X_train['Humidity_n'] = numerics.fit_transform(X_train['Humidity'])
X_train['windy_n'] = numerics.fit_transform(X_train['Windy'])
X_train = X_train.drop(['Outlook', 'Temp', 'Humidity', 'Windy'], axis='columns')
X_train
```

## Modeling

```python
nb = GaussianNB()
nb.fit(X_train, y_train)
```

## Evaluation

```python
nb.score(X_train, y_train)
```

## Tugas
Buat prediksi apakah hari ini bermain golf atau tidak jika diketahui kondisi hari ini adalah Outlook = Sunny, Temperature = Hot, Humidity = Normal, dan Windy = False)

```python
df = pd.read_csv("https://raw.githubusercontent.com/rahmantaufik27/dataset/main/study/weather_nb_dataset.csv")
df
```

```python
# pisahkan data target
X_train = df.drop('Play', axis='columns')
y_train = df['Play']
```

```python
# buat dataframe baru dengan encoding data
numerics = LabelEncoder()
X_train['outlook_n'] = numerics.fit_transform(X_train['Outlook'])
X_train['Temp_n'] = numerics.fit_transform(X_train['Temp'])
X_train['Humidity_n'] = numerics.fit_transform(X_train['Humidity'])
X_train['windy_n'] = numerics.fit_transform(X_train['Windy'])
X_train = X_train.drop(['Outlook', 'Temp', 'Humidity', 'Windy'], axis='columns')
X_train
```

```python
# Melakukan pelatihan model Naive Bayes pada data latih
nb = GaussianNB()
nb.fit(X_train, y_train)
```

Outlook

*   Sunny = 2
*   Rainny = 1
*   Overcast = 0

Temperature

*   Hot = 1
*   Mild = 2
*   Cool = 0

Humidity

*   Normal = 1
*   High = 0

Windy

*   False = 0
*   True = 1











```python
# Membuat prediksi untuk kondisi cuaca hari ini
today_outlook = 2
today_temp = 1
today_humidity = 1
today_windy = 0
prediction = nb.predict([[today_outlook, today_temp, today_humidity, today_windy]])
```

```python
# Menampilkan hasil prediksi
print(f"Probabilitas tidak bermain golf / bermain golf: {nb.predict_proba([[today_outlook, today_temp, today_humidity, today_windy]])}")
if prediction == 0:
    print("Hari ini tidak bermain golf.")
else:
    print("Hari ini bermain golf.")
```

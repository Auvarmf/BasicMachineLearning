```python
import pandas as pd
import numpy as np

# import apriory dari library mlxtend
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
# membuat data tabular yang terdiri dari data transaksi
df = pd.DataFrame(
    {
        "ID Transaksi" : pd.Series([1, 2, 3, 4, 5]),
        "Item" : pd.Series([["Susu", "Roti", "Biskuit", "Mentega"], \
                            ["Susu", "Biskuit", "Gula", "Kopi"], \
                            ["Gula", "Kopi", "Susu"], ["Kopi", "Biskuit", "Susu", "Gula"], \
                            ["Mentega", "Roti", "Kopi"]]),
    }
)
```

## Encoding data transaksi


```python
df = df.explode('Item')
df
```

```python
basket = (df.groupby(["ID Transaksi", "Item"])["Item"].count()\
                                      .unstack().reset_index().fillna(0)\
                                      .set_index("ID Transaksi"))
basket.head()
```

## Algoritma Apriori

Eksekusi algoritma apriori menggunkan librari mlxtend dengan fungsi apriori

```python
# set minimmum support 60%
frequent_itemsets = apriori(basket, min_support=0.6, use_colnames=True)

# print(list(frequent_itemsets))
frequent_itemsets
```

## Apriori Tahap 2
Perhitungan Confidence

```python
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.6)
rules[["antecedents", "consequents", "support", "confidence"]]
```

```python
# Dataset transaksi
dataset = [
    ['susu', 'roti', 'biskuit', 'mentega'],
    ['gula', 'kopi', 'roti', 'biskuit'],
    ['susu', 'gula', 'kopi', 'mentega'],
    ['susu', 'kopi', 'biskuit'],
    ['susu', 'gula', 'mentega'],
    ['roti', 'gula', 'kopi', 'mentega'],
    ['susu', 'roti', 'mentega'],
    ['susu', 'roti', 'gula', 'kopi', 'biskuit', 'mentega'],
    ['susu', 'roti', 'gula', 'mentega'],
    ['susu', 'kopi', 'mentega']
]
```

```python
# Fungsi untuk menghitung itemset kandidat dari itemset sebelumnya
def generate_candidates(prev_itemset, k):
    candidates = []
    n = len(prev_itemset)
    
    for i in range(n):
        for j in range(i+1, n):

            # Mengambil itemset sebelumnya yang memiliki (k-2) item yang sama
            if prev_itemset[i][:k-2] == prev_itemset[j][:k-2]:
              
                # Menggabungkan dua itemset menjadi satu itemset kandidat baru
                candidate = list(set(prev_itemset[i] + prev_itemset[j]))
                candidate.sort()
                
                # Menambahkan itemset kandidat ke dalam list candidates
                if candidate not in candidates:
                    candidates.append(candidate)
    
    return candidates
```

```python
# Fungsi untuk menghitung support count dari setiap itemset dalam dataset
def get_support_count(itemset, dataset):
    count = 0
    for trans in dataset:
        if set(itemset).issubset(set(trans)):
            count += 1
    return count
```

```python
# Fungsi untuk menghitung itemset yang memiliki support count >= minimal support
def get_frequent_itemset(itemsets, min_support, dataset):
    frequent_itemset = {}
    for itemset in itemsets:
        support_count = get_support_count(itemset, dataset)
        support = float(support_count) / len(dataset)
        if support >= min_support:
            frequent_itemset[tuple(itemset)] = support
    return frequent_itemset
```

```python
# Fungsi utama untuk mengeksekusi algoritma Apriori
def apriori(dataset, min_support):
    itemset = [[item] for item in set([item for trans in dataset for item in trans])]
    frequent_itemset = get_frequent_itemset(itemset, min_support, dataset)
    print_frequent_itemset(frequent_itemset)
    
    k = 2
    while len(frequent_itemset) > 0:
        itemset = list(frequent_itemset.keys())
        candidates = generate_candidates(itemset, k)
        frequent_itemset = get_frequent_itemset(candidates, min_support, dataset)
        print_frequent_itemset(frequent_itemset)
        k += 1

# Fungsi untuk menampilkan frequent itemset
def print_frequent_itemset(frequent_itemset):
    for itemset, support in frequent_itemset.items():
        print('{} : {}'.format(list(itemset), round(support, 2)))
```

```python
# Menjalankan algoritma Apriori dengan minimal support 40%
apriori(dataset, 0.4)

# Masih sulit untuk memahami dan membuat algoritma apriori secara manual
```
